```python
# data are here. 
# We use f = mg as the independent variable and mean terminal speeds as dependent.


# added weight in grams
added_weight_g = np.array([50,70,90,110,130,150,170])
# added weight in N
added_weight_N = added_weight_g * 0.0098
total_weight_N = added_weight_N + 0.881

total_weight_N
```

```python
# Inependent variable (air resistance = total weight)
f = total_weight_N
# The error on v is much larger than the error on f.
# Therefore, we ignore the error on f.
```

```python


# Dependent variable, mean terminal speeds
v = np.array([2.29,
2.0,
2.2,
3.0,
2.39,
2.88,
3.36

])

# The error on v is much larger than the error on f.
# Therefore, we ignore the error on f.

# Error on v
err = np.array([0.06,
0.4,
0.2,
0.3,
0.04,
0.05,
0.16


])
```

```python
# first I will try a proportional model, then a quadratic model
```

```python
# The proportional model
# Define model functions  -- the independent variable has to be the first argument
# proportional model v=mf
def model_prop(f,m):
    return m*f
```

```python
# Give curve fit
prop_opt, prop_cov = opt.curve_fit(model_prop, f, v, sigma=err, absolute_sigma=True)

prop_err = np.sqrt(np.diag(prop_cov)) 

print('m = {0:6.4f} ± {1:5.4f} '.format(prop_opt[0], prop_err[0]))
```

```python
# Predicted values
v_prop = model_prop(f, prop_opt[0])

# chi_sqr
def chisq(predicted, observed, errors):
    res = observed-predicted
    norm_res = res / errors 
    return np.sum(norm_res**2)

chi2_prop = chisq(v_prop, v, err)
print("χ2 =", chi2_prop)
```

```python
# Create a scatter plot of the data points
#plt.scatter(x, y, label='Data', color='blue')

# data with error bar
plt.errorbar(f, v, err, fmt='o', capsize = 3, capthick = 1, label="data with error bar")


# Define new array, fModel, for displaying the model
fModel = np.linspace(0,8)  


# Create the regression line
regression_line_prop = prop_opt[0] * fModel

# Plot the regression line
plt.plot(fModel, regression_line_prop, label='Weighted Prop. Regression', color='red')


# Add labels and a legend
plt.xlabel('Air resistance (N)')
plt.ylabel('Mean terminal speed (m/sec)')
plt.legend()

plt.ylim(0,3.6)
plt.xlim(0,2.6)

# add a title
plt.title('v-f Weighted Prop. Regression')

# add some texts
plt.text(1.5,0.7,"v=kf")
plt.text(1.5,0.5,"k=%5.4f\u00b1%5.4f" % (prop_opt[0], prop_err[0]))

plt.text(1.5,0.3,"χ2=%5.3f" %chi2_prop )
plt.text(1.5,0.1,"~χ2=%5.3f" %(chi2_prop/(6-1)) )
# Show the plot
plt.grid(True)
plt.show()
```

```python
# Residuals plot

# Residuals
r_prop = (v- v_prop)/err

# Sets figure size etc.
fig2=plt.figure(figsize=(6, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': '14'})

# Residuals
plt.errorbar(f, r_prop, fmt='o')
plt.axhline(color='r') # 0 line for reference

# Axes label and title
plt.title("Prop. fit residuals")
plt.xlabel('v')
plt.ylabel("Residuals")
plt.show()
```

```python
# The quadratic model
# Define model functions  -- the independent variable has to be the first argument
# quadratic model v=m(f**0.5)
def model_quad(f,m):
    return m*(f**0.5)
```

```python
# Give curve fit
quad_opt, quad_cov = opt.curve_fit(model_quad, f, v, sigma=err, absolute_sigma=True)

quad_err = np.sqrt(np.diag(quad_cov)) 

print('m = {0:6.4f} ± {1:5.4f} '.format(quad_opt[0], quad_err[0]))
```

```python
# Predicted values
v_quad = model_quad(f, quad_opt[0])


chi2_quad = chisq(v_quad, v, err)
print("χ2 =", chi2_quad)
```

```python
# Create a scatter plot of the data points
#plt.scatter(x, y, label='Data', color='blue')

# data with error bar
plt.errorbar(f, v, err, fmt='o', capsize = 3, capthick = 1, label="data with error bar")


# Define new array, fModel, for displaying the model
fModel = np.linspace(0,2.6)  


# Create the regression line
regression_line_quad = quad_opt[0] * (fModel**0.5)

# Plot the regression line
plt.plot(fModel, regression_line_quad, label='Weighted Quad. Regression', color='red')


# Add labels and a legend
plt.xlabel('Air resistance (N)')
plt.ylabel('Mean terminal speed (m/sec)')
plt.legend()

plt.ylim(0,3.6)
plt.xlim(0,2.6)

# add a title
plt.title('v-f Weighted Quad. Regression')

# add some texts
plt.text(1.5,0.7,"v=k(f**0.5)")
plt.text(1.5,0.5,"k=%5.4f\u00b1%5.4f" % (quad_opt[0], quad_err[0]))

plt.text(1.5,0.3,"χ2=%5.3f" %chi2_quad )
plt.text(1.5,0.1,"~χ2=%5.3f" %(chi2_quad/(6-1)) )
# Show the plot
plt.grid(True)
plt.show()
```

```python
# Residuals plot

# Residuals
r_quad = (v- v_quad)/err

# Sets figure size etc.
fig2=plt.figure(figsize=(6, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': '14'})

# Residuals
plt.errorbar(f, r_quad, fmt='o')
plt.axhline(color='r') # 0 line for reference

# Axes label and title
plt.title("Quad. fit residuals")
plt.xlabel('v')
plt.ylabel("Residuals")
plt.show()
```
